```python
# Assignment-1
    # This assignment is for creating Phase-1 of Machine Learning: MODEL CREATION.
    # This is an example project in HR to predict the Salary of an Employee who joins newly.
    # A salary database in .csv file is used as data for this project. .CSV file is Comma seperated value format file. This is similar to excel.
    
    # To handle the salary .csv file pandas library is used. Pandas is a library which handles Database and files.
    
    # As a Step-1 we have to import the pandas library in this file.

# CODE
import pandas as pd 

    # In this line of code pandas library is imported. After writing this code use shift + enter to execute this code.
    # in this step after executing this code, All the functions in the pandas library will be loaded in this python file.
    # As we have to use this name "pandas" frequently in this program, It can be imported as "pd". Now "pd" is a short name of the imported "pandas" library.
    # Hereafter in all places in this program "pd" will be used instead of "pandas". This is created for convenience purpose only.
```

```python
# STEP-1: DATA COLLECTION.

    # In this step, the salary dataset .csv file is loaded in a variable named as "dataset"
    # For doing this read_csv() function is used which is available in the pandas class.
    # The Syntax is pd.read_csv(). Here "pd" is pandas class and read_csv("filename.csv") is the function used to read the .csv file which is available in this folder.
    # the "." is the access operator. Inside the read_csv() function bracket, we have to pass the file name as parameter.

# CODE
dataset=pd.read_csv("Salary_Data.csv")

    # The salaray .csv dataset file should be available in the same folder along with this python file.
    # Only then the program will work. Otherwise Error will be displayed.
    # The spelling of the file name should be correct. Otherwise file not found error will be displayed.
    # Execute this code by pressing shift + Enter.
```

```python
    # To check if the file has been loaded in the variable "dataset", We can run and see the result to know its content.
    # Either the variable name can be typed as it is to know the result. Or, we can use print statement to know the result.

# CODE
dataset # In this code
```

```python
# When you open the Salary.csv dataset file seperately inside the folder, You can see the same data in that file also.
# This ensures that the data from that file is now available in this variable "dataset"

# In the above result the content inside the variable "dataset" has both Input and Output data.
# Input value is the "YearsExperience". This is the "Independent Variable"
# Output value is the "Salary". This is the "Dependent Variable"
```

```python
# STEP-2: DATA CLEANING ( Since this data is already cleaned here we are not doing this step)

# STEP-3: INPUT and OUTPUT splitting inside the dataset.

    # In this step, We will ASSIGN the "Independent variable" or "Input Value" available in the variable "dataset" which is "YearsExperience" to a new variable.
    # The new variable name can be given as "independent" for ease of referring.

# CODE
independent=dataset[["YearsExperience"]]
    # Here in the above dode the syntax to extract only the column "YearsExperience" available in the variable "dataset" is variablename[[""]]
    
    # To check the content inside the newly created independent variable "independent", run and see the result.
independent
```

```python
    # In the same way as above, Create a "dependent variable" or "output" in a new variable.
    # For easy understanding name the new variable as "dependent"

# CODE
dependent=dataset[["Salary"]]
dependent
```

```python
# STEP-3: Splitting the TRAINING SET and the TEST SET.

    # For Splitting the TRAINING SET and TEST SET data, We have to use another libraray.
    # "pandas" can only do the database reletaed functions.
    # For spliting the TRAINING and TEST Set, We have to use "sklearn" Library.
    # "sklearn" library is used for Machine Learning.

# CODE
from sklearn.model_selection import train_test_split
    # In the above code ONLY the function train_test_split is imported under Library.class sklearn.model_selection
    # By doing this only the memory required to store the train_test_split function will be used in this file.
    # If the entier library "sklearn" is imported (like we have imported "pandas" library) then all the classes and functions will be 
    # loaded in this file and the memory consumption will be high.
    # Syntax is : from library.class import function_name

    # After importing the train_test_split function use this function to SPLIT the TRAINING SET and TEST SET data.
    # This is done by passing the pararmeters to this function.
    # Syntax is : train_test_split(independent_variable,dependent_variable,test_size=0.3,random_state=0)
    # In the above syntax, independent_variable,dependent_variable,test_size=0.3,random_state=0 are its parameters.
    # Independent variable is the INPUT which is usually "x"
    # Dependent variable is the OUTPUT which is usually "y"
    # test_size is we are defining the size of the data set for testing purpose. Training set data should be more than test set data.
    # Usually the split ratio can be 70:30. Training set 70% data and test set 30% data can be used.
    # random_stat=0 (will be learnt in later clases)

# Below CODE without assigining to a variable
# CODE: train_test_split(independent,dependent,test_size=0.30,random_state=0)
    # If you run the above code, You will get 4 variables as OUTPUTS splitted into x_training, x_test, y_training, y_test.
    # This cannot be used as it is because the output you get is a complete list.
    # To use this output in further steps of coding, we have to store these 4 outputs in seperate meaningful variable.

# CODE 
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
    # In the above code:
    # In LHS side, position_1 is for input training set data.
    # In LHS side, position_2 is for input test set data.
    # In LHS side, position_3 is for output training set data.
    # In LHS side, positoin_4 is for output test set data.
    # In "x_train" variable only the input training dataSset is stored. This "x_train" variable name is user defined.
    # In "x_test" variable only the input test dataset is stored. This "x_test" variable name is user defined.
    # In "y_train" variable only the output training dataset is stored. This "y_train" variable name is user defined.
    # In "y_test" variable only the output test dataset is stored. This "y_test" variable name is user defined.
    
```

```python
# STEP-4: MODEL CREATION 

    # In this step, we will use "SIMPLE LINEAR REGRESSION ALGORITHM" to solve this problem.
    # "SIMPLE LINEAR REGRESSION ALGORITHM" is used for solving this problem.
    # We will learn how to use "SIMPLE LINEAR REGRESSION ALGORITHM" using PYTHON.

# CODE
from sklearn.linear_model import LinearRegression
    # In the above code, we are importing the class "LinearRegression" from "sklearn.linear_model"
    # import can be a CLASS or a FUNCTION.
    # Now for ease of using this function this "LinearRegression" class is assigned to a new variable "regressor"
    # After assigning, Hereafter all the functions available in the class "LinearRegression" class will be availabel in the variable "regressor"
    # Every time in the coding we need not type "LinearRegression". Instead we can type only "regressor".

# CODE
regressor=LinearRegression() 
    # This CLASS "LinearRegression" is now assigned as non-parameterised class to the variable "regressor"
    # All the FUNCTIONS available in the CLASS "linearRegression" will now be available in the variable "regressor"

    # The above TWO Lines of code is a PROCEDURE.
    # For all types of ALGORITHMS only we have to change the above two lines of codes. This makes it simple.

# Using the "fit" method under Simple Linear Regression, We will create the model.

# CODE
regressor.fit(x_train,y_train)
    # The above code is does the LEARNING or this is the MODEL.
    # This code has the Linear Regression formula: y= mx + c, bias, weight
    # Now the MODEL has been created.
```

```python
# To know the weight and bias, We have to assign a variable.

# CODE
weight=regressor.coef_
print(weight)
    # In the above code, we have used "regressor.coef_" function where the weight is stored
    # variable name "weight" is used to assign the weight value available in the "regressor.fit" function.

# CODE
bias=regressor.intercept_
print(bias)
    # In the above code, we have used "regressor.intercept_" function where the bias is stored
    # variable name "bias" is used to assign the bias value available in the "regressor.fit" function.

```

```python
# STEP-5: MODEL EVALUATION
    # In this step, We are going to check the model working accuracy.
    
    # for this we use the function "regressor.predict()"
    # Already in the dataset, We have test input values and corresponding test output values as reference.
    # Now by using the "predict" function, we will calculate/PREDICT the output value and store it in a new variable name "y_pred"
    # "x_test" values are passed as parameter in the predict function.
    # By doing this step you will get PREDICTED values for the test set values.
    # After predicting the output test values this can be compared with the original test set values to know the accuracy of the predicted results.
    # For checking the accuracy we will use r squared method. If the r square value is close to value 1 then the model is a good model.
    # This model can be saved for using it.

    # 
# CODE:
    # This code PREDICTS the output value based on the test set input value "x_test".
    # The predicted output value is stored in a new variable "y_pred".
y_pred=regressor.predict(x_test)

    # Now we can check the accuracy of the predicted value by using rsqured method.
    # For this we have to import the "r2_score" function from the class "sklearn.metrics" class.
    # after importing this in this file, we can pass the parameter into the "r2_score" function to check the accuracy of the predicted value.
    # 2 Parameters are used for this. 1) The output test value splitted from the original data set
    # and 2) the output value predicted using "regressor.predict" function and stored in the variable "y_pred"
    # The result of the "r2_score" function is store in a new variable r2_score.
    # Execute the code to get the "r_score" result.

# CODE:
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score

    # By executing the above code, We get the value for  r_score=0.9740993407213511.
    # This means that the MODEL is accurate and can be saved for further usage.
```

```python
# STEP-6: SAVING THE CREATED MODEL

    # Model saving is done by using PICKLE
# CODE:
    # in this "pickle" is imported.
import pickle

    # Creating a file name
    # The file name is created with .sav extension.
    # In this step, Only a file name is created but the model is not yet saved here.
# CODE:
filename="finalized_model_linear.sav"

    # SAVING THE MODEL into the created file.
# CODE:
pickle.dump(regressor,open("finalized_model_linear.sav",'wb'))
    # SAVING the Created model "regressor" is done by using "pickle.dump"
    # In the above step the created MODEL "regressor" is saved in the file name "finalized_model_linear.sav"
    # the 'wb' is for WRITE BINARY.
    # After executing this code, You can go to the folder and see this file is created with a .sav file extension.
    # Only if you save the created model, You can use it by giving it to the enduser.

    # LOADING THE MODEL
    # This step is done to check before deployment if the saved model is working properly?
    # A variable is created "loaded_model" and the file is assigned to that by using "pickle.load()"
# CODE:
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
    # After the above step, this model can be checked in realtime.
    # In the above step, the model is loaded in the variable "loaded_model"
    # We can test this loaded model in realtime before deployment.
    # this can be done using "predict"
result=loaded_model.predict([[15]])
    # In the above step, Using predict the USER input is given and the predicted value is stored in the variable "result"
```
